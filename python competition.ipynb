```python
def display_books(books):
    print("\nAvailable Books:\n")
    for title, quantity in books.items():
        print(f"{title} - {quantity} available")

def borrow_book(books):
    title = input("Enter the title of the book you want to borrow: ")
    if title in books and books[title] > 0:
        quantity = int(input(f"Enter the quantity you want to borrow (available: {books[title]}): "))
        if quantity > 0 and quantity <= books[title]:
            books[title] -= quantity
            print(f"You have borrowed {quantity} copy(ies) of {title}.")
        else:
            print("Invalid quantity or not enough copies available.")
    else:
        print(f"{title} is not available for borrowing.")

def return_book(books):
    title = input("Enter the title of the book you want to return: ")
    if title in books:
        quantity = int(input(f"Enter the quantity you want to return: "))
        if quantity > 0:
            books[title] += quantity
            print(f"You have returned {quantity} copy(ies) of {title}.")
        else:
            print("Invalid quantity.")
    else:
        print(f"{title} is not in the inventory.")

def main():
    books = {
        "Book1": 5,
        "Book2": 3,
        "Book3": 7,
        "Book4": 2,
        "Book5": 4
    }

    while True:
        print("\nWelcome to the Basic Library Management System!\n")
        print("1. Display Available Books")
        print("2. Borrow a Book")
        print("3. Return a Book")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            display_books(books)
        elif choice == '2':
            borrow_book(books)
        elif choice == '3':
            return_book(books)
        elif choice == '4':
            print("\nExiting the Library Management System. Thank you!\n")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

if __name__ == "__main__":
    main()
```
